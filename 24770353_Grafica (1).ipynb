```python
#importar Bibliotecas
import numpy as np
import matplotlib.pyplot as plt

#Establecemos el dominio y rango de la funcion
x= np.arange(-np.pi, np.pi, 0.2)
y=np.cos(x)

#graficar la funcion
plt.plot(x,y, "-0", label="$y,cos(x)$")
plt.legend(loc=1)
plt.savefig("GraficaCoseno_pi.png")
plt.grid()
plt.show()
```
