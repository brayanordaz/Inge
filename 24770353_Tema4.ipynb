Elaborar un progaram que permita ingresar el numero de partidos ganados, perdidos y empatados del club de futbol"real albañil" en el torneo apertura 2024, se debe mostrar el puntaje final teniendo en cuenta que por cada partido ganado obtendra 3 puntos, 1 por los empatados y 0 por perder

```python
print("---------------------------------------------")
print("Puntaje final del club de futbol real albañil")
print("---------------------------------------------")

#Entrada
print("Ingrese el numero de partidos ganados")
ganados = int(input())
print("Ingrese el numero de partidos perdidos")
perdidos = int(input())
print("Ingrese el numero de partidos empatados")
empatados =int(input())

ppg=ganados*3
ppp=perdidos*0
ppe=empatados*1

pf=ppg+ppp+ppe
print("El puntaje final del club es: ",pf)
```

Escribir un programa que dado como datos el radio y la altura de un cilindro, calcular e imprimir el area y volumen del  mismo

```python
print("---------------------------------------------")
print("Area y volumen de un cilindro ")
print("---------------------------------------------")

#Entrada
PI=3.1416
print("Ingrese el radio del cilindro")
radio = float(input())
print("Ingrese la altura del cilindro")
altura = float(input())

vol=PI*radio**2*altura
ar=2*PI*radio*(altura+radio)

print("El volumen del cilindro es: ",vol)
print("El area del cilindro es: ",ar)
```

Escribir un pragrama que calcule el perimetro de un triangulo equilatero, teniendo como dato de entrada la altura

```python
print("---------------------------------------------")
print("Perimetro de un triangulo equilatero")
print("---------------------------------------------")

#Entrada
print("Ingrese la altura del triangulo")
altura= float(input())

import math
lado=altura*2/math.sqrt(3)

per=lado*3
print("El perimetro del triangulo es: ",per)
```
