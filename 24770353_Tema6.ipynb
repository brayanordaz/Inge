Escribe un programa que dados 3 numeros muestre el mensaje "IGUALES"si la suma de los dos primeros es igual al otro numero, y el mensaje "DISTINTOS" en caso contrario.

```python
print("Ingrese 3 Numeros: ")
a =int(input("ingrese a: "))
b = int(input("ingrese b: "))
c =int (input("ingrese c: "))

if a+b==c :
  print("iguales")
else:
  print("distintos")
```

Escriba un programa que dada una fecha del año 2000(representada por el dia, mes y año en formato numerico dd/mm/aa) calcule el dia siguiente. Se asume que el mes tiene 30 dias

```python
print("Ingrese fecha: ")
d =int(input("ingrese dia: "))
m = int(input("ingrese mes: "))
a =int (input("ingrese año: "))

if d<30:
  dia=d+1
  mes=m
  año=a
else:
  if m<12:
    dia=1
    mes=m+1
    año=a
  else:
    dia=1
    mes=1
    año=a+1

print("El dia siguiente es: ",dia,"/",mes,"/",año)


```
